```python
import os
import json
import requests
from dotenv import load_dotenv

load_dotenv()


def get_api_key():
    api_key = os.getenv("OMDB_API_KEY")
    if not api_key:
        raise ValueError("Ключ OMDB_API_KEY не найден. Проверьте файл .env")
    return api_key


def get_movie_title():
    movie_title = input("Введите фильм для поиска: ").strip()
    if not movie_title:
        raise ValueError("Название фильма не должно быть пустым")
    return movie_title


def fetch_movie(movie_title, api_key):
    response = requests.get(
        "http://www.omdbapi.com/",
        params={
            "t": movie_title,
            "apikey": api_key,
        },
        timeout=10,
    )
    response.raise_for_status()
    return response.json()


def print_movie_info(movie_data):
    if movie_data.get("Response") == "False":
        print("Фильм не найден:", movie_data.get("Error", "неизвестная ошибка"))
        return

    print(json.dumps(movie_data, ensure_ascii=False, indent=2))

    movie_title = movie_data.get("Title")
    year = movie_data.get("Year")
    genre = movie_data.get("Genre")
    director = movie_data.get("Director")
    actors = movie_data.get("Actors")
    imdb_rating = movie_data.get("imdbRating")

    print(f"Название: {movie_title}")
    print(f"Год: {year}")
    print(f"Жанр: {genre}")
    print(f"Режиссёр: {director}")
    print(f"Актёры: {actors}")
    print(f"Рейтинг IMDb: {imdb_rating}")
    print(f"Фильм '{movie_title}' ({year}) - жанр: {genre}. Режиссёр: {director}.")


api_key = get_api_key()
movie_title = get_movie_title()
movie_data = fetch_movie(movie_title, api_key)
print_movie_info(movie_data)
```
